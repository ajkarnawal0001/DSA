2. Doubly Linked List Stack
Insertion (push) at Tail (O(1))
Deletion (pop) from Tail (O(1))
Allows backward traversal (supports more operations)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class StackUsingDLL:
    def __init__(self):
        self.tail = None
        self.size = 0

    def push(self, data):  # O(1)
        newNode = Node(data)
        if self.tail is None:
            self.tail = newNode
        else:
            self.tail.next = newNode
            newNode.prev = self.tail
            self.tail = newNode
        self.size += 1

    def pop(self):  # O(1)
        if self.tail is None:
            return "Stack Underflow"
        data = self.tail.data
        self.tail = self.tail.prev
        if self.tail:
            self.tail.next = None
        self.size -= 1
        return data

    def peek(self):  # O(1)
        if self.tail is None:
            return "Stack is empty"
        return self.tail.data

    def is_empty(self):  # O(1)
        return self.size == 0

    def get_size(self):  # O(1)
        return self.size
```

3. Circular Linked List Stack
Last node points to the first node
Insertion (push) and Deletion (pop) at Head (O(1))
Efficient for circular memory usage

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class StackUsingCLL:
    def __init__(self):
        self.head = None
        self.size = 0

    def push(self, data):  # O(1)
        newNode = Node(data)
        if self.head is None:
            self.head = newNode
            newNode.next = self.head  # Circular link
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = newNode
            newNode.next = self.head
            self.head = newNode  # New node becomes the head
        self.size += 1

    def pop(self):  # O(1)
        if self.head is None:
            return "Stack Underflow"
        
        if self.head.next == self.head:  # Only one node
            data = self.head.data
            self.head = None
        else:
            temp = self.head
            while temp.next != self.head:
                prev = temp
                temp = temp.next
            prev.next = self.head  # Maintain circular link
            data = temp.data
        self.size -= 1
        return data

    def peek(self):  # O(1)
        if self.head is None:
            return "Stack is empty"
        return self.head.data

    def is_empty(self):  # O(1)
        return self.size == 0

    def get_size(self):  # O(1)
        return self.size
```

4. Stack with Min Operation in O(1)
Keeps track of the minimum element
Push and Pop in O(1)
Uses an additional stack to track min values

```python
class Node:
    def __init__(self, data, min_val):
        self.data = data
        self.min_val = min_val
        self.next = None

class StackWithMin:
    def __init__(self):
        self.head = None
        self.size = 0

    def push(self, data):  # O(1)
        if self.head is None:
            newNode = Node(data, data)  # First element, min = data
        else:
            newNode = Node(data, min(data, self.head.min_val))
            newNode.next = self.head
        self.head = newNode
        self.size += 1

    def pop(self):  # O(1)
        if self.head is None:
            return "Stack Underflow"
        data = self.head.data
        self.head = self.head.next
        self.size -= 1
        return data

    def get_min(self):  # O(1)
        if self.head is None:
            return "Stack is empty"
        return self.head.min_val
```
